```python
# imports
import requests, subprocess, os
from time import sleep

from papers import FetchPMIDS, PMID2PMCID, FetchLiteraturePMID, CleanXml, DownloadPaper, WritePaper, SearchPubmed
```

```python
accession='P04529'
```

```python
pmids=FetchPMIDS(accession, 'protein')
pmids
```

```python
papers=[]
for pmid in pmids:
    pmcid=PMID2PMCID(pmid)
    if pmcid:
        papers.append(pmcid)
    if not pmcid:
        papers.append(pmid)

papers
```

```python
os.makedirs(f'downloads/{accession}', exist_ok=True)
for paper in papers:
    if paper.startswith('PMC'):
        DownloadPaper(paper, f'downloads/{accession}/{paper}.txt')
    else:
        p=FetchLiteraturePMID(paper)
        p=CleanXml(p)
        WritePaper(p, f'downloads/{accession}/{paper}.txt')
```
