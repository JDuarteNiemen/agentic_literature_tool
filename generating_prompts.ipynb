```python
# Imports
from pydantic import BaseModel, create_model
import requests
import numpy as np
from langchain_ollama import ChatOllama
import json
```

```python
# Set model
model = 'gemma4:e4b'

# Set structures output for returning a prompt
class PromptOutput(BaseModel):
    prompt: str

```

```python
llm=ChatOllama(model=model,
               reasoning=False,
               temperature=0.0,
               num_predict=4096)

llm.with_structured_output(PromptOutput)
```

```python
# Set the topic
topic = 'thermal adaptation'
```

```python
def initial_prompt(topic):
    return f"""I would like you to create a prompt that can be used in downstream LLM calls to obtain information useful in a summary around the given topic.
Topic = {topic}
Just the prompt no other text.
For the paper text put the word 'paper' in curly braces so it can be used as an f-string. Put this as the end and mark as 'Paper: {{paper}}'
"""
```

```python
response = llm.invoke(initial_prompt(topic))

summary_prompt=response.content
print(f'Summary prompt:\n{summary_prompt}')
```

```python
topics=['thermal adaptation', 'host species', 'methodology', 'habitat', 'migration patterns', 'population density', 'susceptibility/resistance']
prompts={}
for topic in topics:
    resp=llm.invoke(initial_prompt(topic))
    prompt=resp.content
    prompts[topic]=prompt
```
